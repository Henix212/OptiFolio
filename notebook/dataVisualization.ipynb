# Data Visualization

This notebook provides exploratory visualizations of financial data: prices, volumes, returns, and volatility. The goal is to get a quick overview of the time series to spot trends, anomalies, and prepare further analysis.

## Prerequisites

Libraries used: `pandas` and `matplotlib`.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
returns = pd.read_csv("../data/returns/data_returns.csv", index_col="Date", parse_dates=True)

returns = returns * 100

returns.plot(subplots=True, title="Returns", figsize=(20,10))
```

```python
def plot_bb(bb, ticker):
    bb_ticker = bb.xs(ticker, level=1, axis=1)

    bb_ticker.plot(
        title=f"Bollinger Bands  {ticker}",
        figsize=(20, 6)
    )

bb_lower = pd.read_csv("../data/indicators/bb_lower.csv", index_col="Date", parse_dates=True)
bb_middle = pd.read_csv("../data/indicators/bb_middle.csv", index_col="Date", parse_dates=True)
bb_upper = pd.read_csv("../data/indicators/bb_upper.csv", index_col="Date", parse_dates=True)

bb = pd.concat(
    {
        "Lower": bb_lower,
        "Middle": bb_middle,
        "Upper": bb_upper,
    },
    axis=1
)

plt.figure()
plot_bb(bb,'USO')
plot_bb(bb,'GLD')
plot_bb(bb,'URTH')
```
