# Data Visualization

This notebook provides exploratory visualizations of financial data: prices, volumes, returns, and volatility. The goal is to get a quick overview of the time series to spot trends, anomalies, and prepare further analysis.

## Prerequisites

Libraries used: `pandas` and `matplotlib`.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Load data

Read CSV files for each dataset and ensure the `Date` column is parsed as a `datetime` type and set as the DataFrame index so the x-axis displays dates properly.

```python
prices = pd.read_csv("../data/cleaned/prices/prices.csv")
volumes = pd.read_csv("../data/cleaned/prices/volumes.csv")
returns = pd.read_csv("../data/cleaned/returns/returns.csv")
volatility = pd.read_csv("../data/cleaned/volatility/volatility.csv")
```

### Prices

For each ticker: convert `Date` to `datetime` if necessary, set it as the index, sort chronologically, and plot each ticker in its own subplot. Rotate x-axis labels and apply `tight_layout` to improve readability.

```python
prices["Date"] = pd.to_datetime(prices['Date'], format='%Y-%m-%d') 
prices.set_index('Date', inplace=True)
prices.sort_index(inplace=True)

prices.plot(subplots=True, figsize=(30, 15), xlabel="Date", ylabel="Prices"); plt.legend(loc='best')
```

### Volumes

Volume charts help detect spikes in activity, which often precede or confirm price movements.

```python
volumes["Date"] = pd.to_datetime(volumes['Date'], format='%Y-%m-%d') 
volumes.set_index('Date', inplace=True)
volumes.sort_index(inplace=True)

volumes.plot(subplots=True, figsize=(30, 15)); plt.legend(loc='best')
```

### Returns

Returns allow analysis of performance and risk. They are sensitive to extreme values

```python
returns["Date"] = pd.to_datetime(returns['Date'], format='%Y-%m-%d') 
returns.set_index('Date', inplace=True)
returns.sort_index(inplace=True)

returns.plot(subplots=True, figsize=(30, 15), use_index = True)
```

### Volatility

Volatility measures variability of returns and serves as a risk indicator. Visualizing historical volatility helps identify calm vs. stressed periods.

```python
volatility["Date"] = pd.to_datetime(volatility['Date'], format='%Y-%m-%d') 
volatility.set_index('Date', inplace=True)
volatility.sort_index(inplace=True)

volatility.plot(subplots=True, figsize=(30, 15)); plt.legend(loc='best')
```
