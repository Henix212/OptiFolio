# Visualisation des features
Objectif: Charger et visualiser les features.


```python
import matplotlib.pyplot as plt
import pandas as pd

import os
```

```python
volatility_path = "data/features/volatility"
returns_path = "data/features/returns"
```

```python
long_volatility = pd.read_csv(os.path.join("..",volatility_path,"long_ratio.csv"))
short_volatility = pd.read_csv(os.path.join("..",volatility_path,"short_ratio.csv"))
vol_ratio = pd.read_csv(os.path.join("..",volatility_path,"vol_ratio.csv"))
zscore = pd.read_csv(os.path.join("..",volatility_path,"zscore.csv"))
returns = pd.read_csv(os.path.join("..",returns_path,"norm_returns.csv"))
```

```python
returns["Date"] = pd.to_datetime(returns['Date'], format='%Y-%m-%d') 
returns.set_index('Date', inplace=True)
returns.sort_index(inplace=True)

returns.plot(subplots=True, figsize=(30, 15), xlabel="Date", ylabel="Prices"); plt.legend(loc='best')
```

```python
long_volatility["Date"] = pd.to_datetime(long_volatility['Date'], format='%Y-%m-%d') 
long_volatility.set_index('Date', inplace=True)
long_volatility.sort_index(inplace=True)

long_volatility.plot(subplots=True, figsize=(30, 15), xlabel="Date", ylabel="Volatility"); plt.legend(loc='best')
```

```python
short_volatility["Date"] = pd.to_datetime(short_volatility['Date'], format='%Y-%m-%d') 
short_volatility.set_index('Date', inplace=True)
short_volatility.sort_index(inplace=True)

short_volatility.plot(subplots=True, figsize=(30, 15), xlabel="Date", ylabel="Volatility"); plt.legend(loc='best')
```

```python
vol_ratio["Date"] = pd.to_datetime(vol_ratio['Date'], format='%Y-%m-%d') 
vol_ratio.set_index('Date', inplace=True)
vol_ratio.sort_index(inplace=True)

vol_ratio.plot(subplots=True, figsize=(30, 15), xlabel="Date", ylabel="Volatility"); plt.legend(loc='best')
```

```python
zscore["Date"] = pd.to_datetime(zscore['Date'], format='%Y-%m-%d') 
zscore.set_index('Date', inplace=True)
zscore.sort_index(inplace=True)

zscore.plot(subplots=True, figsize=(30, 15), xlabel="Date", ylabel="Volatility"); plt.legend(loc='best')
```
