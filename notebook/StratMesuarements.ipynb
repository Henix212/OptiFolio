```python
df = pd.read_csv("../model/results/Portfolio_evo.csv")

df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

df['daily_return'] = df['Portfolio_Value'].pct_change()

mean_daily = df['daily_return'].mean() * 252
vol_annual = df['daily_return'].std() * np.sqrt(252)
sharpe = mean_daily / vol_annual

summary = pd.DataFrame({
    'Annual return': [mean_daily * 100], 
    'Volatility': [vol_annual * 100],
    'Sharpe': [sharpe],
}, index=['AI Strategy'])

print(summary)
```
