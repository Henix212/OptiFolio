```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Categorical
from torch.utils.data import Dataset

import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.cuda.is_available()
```

### Definition of both the Actor and the Critic network structure

```python
def A2C(n_samples,n_action,critic_lr,actor_lr):
    critic_layers = [
        nn.Linear(n_samples,32),
        nn.ReLU(),
        nn.Linear(32,32),
        nn.ReLU(),
        # Only one output because the critc network only as to give an estimation of V(s)
        nn.Linear(32,1) 
    ]

    actor_layers = [
        nn.Linear(n_samples,32),
        nn.ReLU(),
        nn.Linear(32,32),
        nn.ReLU(),
        nn.Linear(32, n_action)
    ]

    critic_network = nn.Sequential(*critic_layers).to(device)
    actor_network = nn.Sequential(*actor_layers).to(device)

    critic_optim = optim.RMSprop(critic_network.parameters(), lr=critic_lr)
    actor_optim = optim.RMSprop(actor_network.parameters(), lr=actor_lr)
```
